```python
import pandas as pd
df = pd.read_csv("Capital_Bike_Sharing_Project.csv")
df = df.iloc[:, :-1]
```

```python
df.info()
```

```python
df = df.astype({"ride_id": "string", 
                "rideable_type":"category",
                "started_at":"datetime64",
                "ended_at":"datetime64",
                "start_station_name":"category",
                "start_station_id":"category",
                "end_station_name":"category",
                "end_station_id":"category",
                "member_casual":"category",
                "start_lat":"float32",
                "start_lng":"float32",
                "end_lat":"float32",
                "end_lng":"float32",
                "member_casual":"category"
                })
```
